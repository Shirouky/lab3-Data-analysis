```python
import pandas as pd

def import_json():
  FILE_NAME = "beads.json"
  df = pd.read_json(FILE_NAME)
  return df

def save_xlsx(df):
  df.to_excel("DataFrame beads.xlsx", index=False)
```

```python
import random

def fill_null(df):
  for repeat in range(random.randint(5, 30)):
      start = random.randint(0, len(df) - 3)
      end = start + random.choice([2, 3])
      df.loc[start:end, "color"] = None
  return df

def fill_color(df):
  mode_value = df["color"].mode()
  df["color"] = df["color"].fillna(mode_value)
  return df

def fill_price(df):
    df_filled = df.copy()
    df_filled['sale'] = df_filled['sale'].fillna(0)
    df_filled['old_price'] = df_filled['old_price'].fillna(df_filled['price'])

    return df_filled
```

```python
def drop(df, rows_mask):
  return df[~rows_mask]

def drop_absurd(df):
  df = df.drop('id', axis=1)
  df = df.drop_duplicates()

  df = df[~df.isnull()["code"]]
  df = df[~(df["price"] < 10)]
  df = df[~(df["type"] == "TWIN")]
  df = df[~(df["weight"] != 50)]
  df = df[~(df["color"] == "")]
  df = df[~(df["quantity"] == "много")]
  return df
```

```python
def strip(column):
  col_new = column.replace(" ", "")
  return col_new

def replace_code(column):
  temp = column.split('/')
  col_new = temp[1] if len(temp) > 1 else temp[0]
  return col_new
```

```python
def factorize(df):
  df["color_category"] = pd.factorize(df["color"])[0] + 1
  df["type_category"] = pd.factorize(df["type"])[0] + 1
  df["code_category"] = pd.factorize(df["code"].str[:2])[0] + 1
  df["color_code_category"] = pd.factorize(df["code"].str[2:5])[0] + 1
  return df

def sort(df):
  df = df.sort_values(by=["color_code_category", "code_category", "color_category", "type_category"], ascending=[False, False, False, False])
  return df
```

```python
def slice_by_mean(column="price"):
  sliced_df = df[df[column] > df[column].mean()]
  return sliced_df

def concat_name(df):
  code = str(df["code"])
  price = str(df["price"])
  col_new = f"{code}/{price}"
  return col_new
```

```python
df = import_json()

df = fill_null(df)
df = fill_color(df)
df = fill_price(df)

df = drop_absurd(df)
df["code"] = df["code"].apply(strip)
df["code"] = df["code"].apply(replace_code)

df = factorize(df)
df = sort(df)

df = slice_by_mean()
df["concat_price"] = df.apply(concat_name, axis=1)
save_xlsx(df)
df
```
